### 2. 데이터 전처리

### 2-1 결측치처리

- NaN값 개수를 확인
- NaN값개수가 너무많으면 그 컬럼을 삭제
- NaN값개수가 적당하면 행을 삭제

```python
# 컬럼별 결측치 개수 확인하기
df_12.isnull().sum()
```

```python
# 결측치가 많은 매출액 성장률(8465개)과 매출채권 회전율(4237개) 제외
df_12 = df_12.drop(['매출액성장률','매출채권회전율'],axis=1)

# 나머지 결측치제거 
df_12 = df_12.dropna().reset_index(drop=True) # index초기화
```

```python
# 결측지 제거 후 데이터 확인
df_12
```

```python
# 결측치 개수 확인
df_12.isnull().sum()
```

### 2-2 inf값 처리

- inf값도 결측치와 마찬가지로 데이터 대부분이면 그 컬럼값을 삭제
- 그런 경우가 아니라면 행값을 삭제

```python
import numpy as np

# 각 컬럼별 inf 값의 개수 확인
inf_count_per_column = (df_12 == np.inf).sum()

print(inf_count_per_column)
```

```python
# inf값이 4000개 이상 컬럼인 매출액 성장률과 매출채권 회전율 제외
df_12 = df_12.drop(['매출총이익률','영업이익률'],axis=1)

# inf 값이 있는 행 제거
df_12 = df_12[~df_12.isin([np.inf, -np.inf]).any(axis=1)]
df_12.reset_index(drop=True, inplace=True)

df_12
```

```python
# 이상치 처리전 데이터 저장
df_12.to_csv("./2. 전처리완료.csv", index=None) # (index=None : 인덱스를 제외하고 데이터만 저장)
```

### 2-3 이상치 제거 

### 1) 데이터 분할
- 2015~2020년은 train 데이터, 2021~2022년은 test 데이터로 분리

- box plot, qq플롯으로 데이터 분포 확인
- 이상치제거를 위해 윈저라이징 진행
- 윈저라이징 전 train 기간과 test기간을 나누어서 train 기간에만 적용할 수 있도록 진행

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np

# 시각화 라이브러리
import matplotlib.pyplot as plt
import seaborn as sns

# 한글폰트 출력
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 경고무시
import warnings 
warnings.filterwarnings('ignore')

# 통계 라이브러리리
from scipy.stats import norm
from scipy import stats
from statsmodels.formula.api import ols
from scipy.stats import kstest
```

```python
# 2015~2020년은 train 데이터, 2021~2022년은 test 데이터로 분리
train_df = df_12[df_12['결산년도'] <= 2020]
test_df = df_12[df_12['결산년도'] >= 2021]

# 확인
print("Train data shape:", train_df.shape)
print("Test data shape:", test_df.shape)
```

```python
# 전처리 후 피처후보
columns = train_df.columns[4:19]
columns
```

### 2) histogram

```python
# subplot설정
col_n = 5
row_n = 3
# 출력 크기 설정
fig, ax = plt.subplots(ncols=col_n, nrows=row_n, figsize=(30,row_n*3))

## histogram
for i,col in enumerate(columns):
    sns.distplot(train_df[col], bins=20, ax=ax[int(i/col_n),int(i%col_n)])
```

### 3) QQ Plot (Quantile-Quantile Plot)
- 두 데이터 집합의 분포를 비교하기 위해 사용되는 통계적 그래프 
- 특히 데이터가 특정 이론적 분포(예: 정규 분포)를 따르는지 확인하거나, 두 데이터 집합이 동일한 분포를 따르는지 검토하는 데 유용
- 해석 방법
- 대각선과 일치 : 데이터가 비교하는 분포를 잘 따르고 있음을 나타냅니다.
- 예: 데이터가 정규 분포를 따른다면 점들이 대각선 근처에 위치합니다.
- 대각선에서 벗어남 : 데이터가 해당 분포를 따르지 않음을 시사합니다.
- 예: 꼬리가 두꺼운 분포를 가지는 데이터는 끝 부분에서 대각선에서 벗어나는 패턴을 보입니다.

```python
# subplot설정
col_n = 5
row_n = 3
# 출력 크기 설정
fig, ax = plt.subplots(ncols=col_n, nrows=row_n, figsize=(30,row_n*3))

# qqplot
for i,col in enumerate(columns):
    stats.probplot(train_df[col],dist=stats.norm, plot=ax[int(i/col_n),int(i%col_n)])
```

### 4) box plot
1. 최소값 (Minimum)
보통 하단 "수염(whisker)"의 끝으로 표시됩니다.
2. 1사분위수 (Q1, 1st Quartile) : 데이터의 하위 25%가 포함된 값입니다.
박스의 하단 경계로 표시됩니다.
3. 중앙값 (Median) : 데이터셋의 중간 값으로, 상위 50%와 하위 50%를 나누는 기준입니다.
4. 3사분위수 (Q3, 3rd Quartile) : 데이터의 상위 75%가 포함된 값입니다.
박스의 상단 경계로 표시됩니다.
5. 최대값 (Maximum) : 데이터셋에서 가장 큰 값입니다.
보통 상단 "수염(whisker)"의 끝으로 표시됩니다.
6. IQR (Interquartile Range, 사분위 범위)
QR=Q3−Q1 데이터의 중간 50% 범위를 나타냅니다.
7. 이상치 (Outliers)
상기 기준(IQR)을 초과하는 값으로, 박스 플롯에서는 보통 점(dot)으로 표시됩니다.

```python
# subplot설정
col_n = 5
row_n = 3
# 출력 크기 설정
fig, ax = plt.subplots(ncols=col_n, nrows=row_n, figsize=(30,row_n*3))

# box plot
for i,col in enumerate(columns):
    train_df.boxplot(column=col,ax=ax[int(i/col_n),int(i%col_n)])
```

### 5) 윈저라이징을 이용한 이상치 제거
윈저라이징(Winsorizing) 방법 : 이상값의 영향을 완화하여 데이터를 정규화하는 데 사용되는 강력한 통계 방법

```python
# 윈저라이징 라이브러리
from scipy.stats.mstats import winsorize

# 재무비율 컬럼에 대해 양쪽 1%씩 총2% 윈저화 적용
train_df_winsor = train_df.apply(lambda columns: winsorize(columns, limits=[0.01, 0.01]))

train_df_winsor
```

```python
# subplot설정
col_n = 5
row_n = 3
fig, ax = plt.subplots(ncols=col_n, nrows=row_n, figsize=(30,row_n*3))

## histogram
for i,col in enumerate(columns):
    sns.distplot(train_df_winsor[col], bins=20, ax=ax[int(i/col_n),int(i%col_n)])
```

```python
# subplot설정
col_n = 5
row_n = 3
# 출력 크기 설정
fig, ax = plt.subplots(ncols=col_n, nrows=row_n, figsize=(30,row_n*3))

# qqplot
for i,col in enumerate(columns):
    stats.probplot(train_df[col],dist=stats.norm, plot=ax[int(i/col_n),int(i%col_n)])
```

```python
# subplot설정
col_n = 5
row_n = 3
# 출력 크기 설정
fig, ax = plt.subplots(ncols=col_n, nrows=row_n, figsize=(30,row_n*3))

# box plot
for i,col in enumerate(columns):
    train_df.boxplot(column=col,ax=ax[int(i/col_n),int(i%col_n)])
```

```python
# 데이터 저장
train_df_winsor.to_csv("./2. train_df.csv",index=None) # 이상치 처리된 train data (index=None : 인덱스를 제외하고 데이터만 저장)
test_df.to_csv("./2. test_df.csv",index=None) # test data (index=None : 인덱스를 제외하고 데이터만 저장)
```
