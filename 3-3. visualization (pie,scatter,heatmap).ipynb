## Pie Chart

```python
# 질적자료(범주형)에 대한 그래프
# 파이 챠트 : 원을 그린후 그 원에 각 계급의 상대도수에 대응하는 면적 또는 부분으로 나눈다.

# 도수(frequency) : 각 범주에 속하는 관측값의 개수 (건수)
# 상대도수(relative frequency) : 도수를 전체 개수로 나눈 비율==> autopct
# 도수분포표(frequency table) : 범주형 자료에서 범주와 그 범주에 대응하는 도수와 상대도수를 나열하는 표

# plt.pie() :  파이 모양의 차트를 표현 가능
# autopct : 각 파이 각각에 문자열을 출력하는 것을 설정

```

```python
import matplotlib.pyplot as plt
```

```python
# 아래 리스트 데이터에 대하여 파이 챠트

x = ['item1', 'item2', 'item3', 'item4']
y = [30, 100, 55, 10]

# ‘%.2f’는 소수점 두번째만 출력하라는 의미
# ‘%%’는 문자 ‘%’를 표현하는 것
# explode :  # explode : 조각내기,   보통 강조할 때 쓰이는 것으로 각 파이를 분리하여 표현하는 정도


plt.pie(y, labels=x, autopct='%.2f%%', explode=(0, 0.1, 0, 0))
plt.title('Bar Chart')
plt.show()
```

```python
x = ['item1', 'item2', 'item3', 'item4']
y = [30, 100, 55, 10]

# ‘%.2f’는 소수점 두번째만 출력하라는 의미
# ‘%%’는 문자 ‘%’를 표현하는 것
# explode :  # explode : 조각내기,   보통 강조할 때 쓰이는 것으로 각 파이를 분리하여 표현하는 정도

plt.pie(y, labels=x, autopct='%.f%%', explode=(0, 0.1, 0, 0))
plt.title('Bar Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# 상관계수 분포 확인을 위한 데이터 구성
columns = ['X' + str(x).zfill(2) for x in range(10)]
 
data = {}
for k in columns:
    data[k] = [0]*20
 
map_df = pd.DataFrame(data)
map_df.index = ['Y' + str(x).zfill(2) for x in range(20)]
 
for i in range(map_df.shape[0]):
    for j in range(map_df.shape[1]):
        map_df.iloc[i,j] = np.random.randn(1)
```

```python
fig = plt.figure(figsize=(8,8))
fig.set_facecolor('white')
 
plt.pcolor(map_df.values) ## matplotlib에서 히트맵을 그려주기 위하여 pcolor 함수 이용
plt.xticks(range(len(map_df.columns)),map_df.columns) ## x축 눈금 생성
plt.yticks(range(len(map_df.index)), map_df.index) ## y축 눈금 생성
 
plt.colorbar()  # Heat Map에 pcolor 함수와 함께 컬러바를 표시해야 함
plt.show()
```

## violin plots

```python
# 일변량, 연속형 데이터의 분포를 설명하기 위해 사용되는 그래프
# 대부분의 내용이 박스 플롯 정보 포함
```

```python
# 일변량(1개) 연속형 데이터 :  히스토그램(Histogram) , 커널 밀도 곡선(Kernel Density Curve) 
# 박스 그래프(Box Plot) ,# 바이올린 그래프(Violin Plot)

#일변량(1개) 범주형 데이터 : 막대 그래프(Bar Chart),원 그래프(Pie Chart)

# 다변량(2개 이상)연속형 데이터  :  선 그래프(Line Chart), 산점도(Scatter)

# 다변량 범주형 데이터 : 히트맵(Heatmap)
#seaborn 패키지에 내장되어 있는 tips 라는 DataFrame 을 사용
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.violin(df, y="tip", x="smoker", color="sex", box=True, points="all", hover_data=df.columns)
fig.show()
```

## box_plot 

```python
# 분포 형태, 퍼짐정도, 이상치 여부 등을 시각화하고, 하나의 그룹 혹은 여러개의 그룹 간 비교하는데 유용
# 최소값(min), 제 1사분위수(Q1), 중앙값(Q2, median), 제 3사분위수(Q3), 최대값(max) 의 요약통계량을 계산하는 것에서 시작
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = [10, 6]
%matplotlib inline
```

```python
# seaborn 패키지에 내장되어 있는 'tips'  데이터셋 사용
```

```python
# loading 'tips' dataset

tips = sns.load_dataset('tips')
```

```python
# 성별과 날짜로 묶기

tips.groupby(['sex', 'day']).size()
```

```python
# boxplot() 함수를 사용
# default 세팅으로 상자 그림

plt.boxplot(tips['tip'])
plt.show()
```

```python
# 상자 그림의 이상치(outlier) 모양과 색깔, 제목(title), X축 이름(X label)을 설정
# outlier symbol, title, xlabel

plt.boxplot(tips['tip'], sym="bo")
plt.title('box of tip')
plt.xticks([1], ['tip'])
plt.show()
```

```python
# 한 개의 축에 여러 개의 연속형 변수(total_bill, tip)에 대한 여러 개의 상자 그림

fig, ax = plt.subplots()
ax.boxplot([tips['total_bill'], tips['tip']], sym="b*")
plt.title('Multiple box plots of tips on one Axes')
plt.xticks([1, 2],['total_bill', 'tip'])
plt.show()
```

```python
 # seaborn 으로 여러개 그룹에 대한 상자 그림 그리기
```

```python
# 요일(day) 그룹별러 팁(tip)에 대한 상자 그림 비교

import seaborn as sns
sns.boxplot(x="day",y="tip",data=tips)
plt.show()
```

```python
# pandas의 DataFrame에 df.boxplot() 함수를 사용하여 상자 그림

tips.boxplot(column=["tip"], by=["day", "sex"])
plt.show()
```
