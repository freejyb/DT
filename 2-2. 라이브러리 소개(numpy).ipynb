### 배열이란?

```python
# 데이터 처를 효율적으로 하기 위한 자료구조
# 배열 안에는 숫자나 문자열  모두를 담을 수 있다. : 배열 안에서는 동일 종류의 데이터만 담을 수 있다.
# - 리스트는 다른 유형의 데이터를 담을 수 있지만 배열은 불가
# 리스트 : [1,2,3] , [안산, 대학교, 금융], [1,2, 안산, 4]
# 배열 : ([1,2,3]) , ([안산, 대학교, 금융])
```

```python
# 배열의 종류(차원: Dimension  ) : 1차원, 2차원, 3차원
# n차원의 배열 (ndarray)
# 1) 1차원 배열 : 삼성전자  주가 ([61000,59000,56500])
# 리스트와 유사하나 , 연산 방법의 차이로 속도가 빠르다.
# 2) 2차원 배열 : 삼성전자 주가, 네이버 주가 ( [61000,59000,56500], [ 200000, 210000, 2250000])
# 3) 3차원 배열 : 2차원 배열이 수개 모여 있는 형태 
```

### 배열 만들기 

```python
# 배열을 만들기 위한 함수
# 1) np.array  : 초기 값을 지정하여 배열 만들기
# 2) np.arrange : 순차적인 숫자 이용 배열 만들기
# 3) np. rand : 랜덤하게 숫자 배열 만들기
```

### np.array

```python
# 만드는 방법
# np.array ([나열된 데이터 객체])
```

```python
# 넘파이 불러오기
import numpy as np
# np.array ([나열된 데이터 객체])
# stock1 리스트 데이터를 배열 형태로 변환

stock1=[15000, 14500, 16000, 17000]
arr1=np.array([stock1])
arr1
```

### np.arange 

```python
# 기본 내장함수 range 와 유사함 : 연속된 숫자로 배열 만들기
# 코드 방법 :  np.arrange([])
```

```python
import numpy as np
np.arange(10)
```

```python
# np.arange(시작번호, 끝번호)
np.arange(1, 10)
```

```python
# np.arange(시작번호, 끝번호, 건너뛰는 수)
np.arange(1, 20, 2)
```

### 배열 탐색하기

```python
# 배열의 타입을 확인하는 경우
# shape : 배열 형태 확인하는 경우
# dtype : 배열 데이터 타입 확인하는 경우
# astype : 배열의 데이터 타입 변경하기
# 인덱스 범위 지정 : 배열 안 데이터 일부 만 보는 경우
```

#### shape : 배열 형태 확인하는 경우

```python
arr=np.arange(10)
arr
```

```python
# arr 에 대하여 배열 형태 확인하는 경우
arr.shape
# 결과 10개 값인 1차원 배열
```

#### dtype : 배열 데이터 타입 확인하는 경우 

```python
# arr 에 대하여 배열에 포함된 데이터 타입 확인하는 경우
arr.dtype
# 결과 정수 값으로 이루어 짐.
```

#### astype : 배열의 데이터 타입 변경하기

```python
# 문자형으로 데이터 타입을 변경하는 경우
arr1=arr.astype(str)
# 문자형으로 변경됨 : '' 안에 숫자가 있다.
arr1
```

#### 인덱스 범위 지정

```python
# 인덱스 범위 지정 : 배열 안 데이터 일부 만 보는 경우
```

```python
arr
```

```python
# 4부터 7번 (-1) 까지
arr[4:7]
```

```python
# 값을 할당하기
# 인덱스 2 다음에 20 할당
arr[2]=20
arr
```
